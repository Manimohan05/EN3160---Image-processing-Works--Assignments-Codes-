Increasing the Brightness Using OpenCV

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('images\Sanga-ICC-Hall-of-Fame.jpg', cv.IMREAD_GRAYSCALE)
# Add 100 to each pixel value of the grayscale image (imgb = img + 100)
imgb = img + 100

# Alternatively, use cv.add() to add 100 to each pixel value of the grayscale image (imgc = cv.add(img, 100))
imgc = cv.add(img, 100)

# Create a figure and a set of subplots with 1 row and 3 columns
f, ax = plt.subplots(1, 3)

# Display the original image in the first subplot
ax[0].imshow(img, cmap="gray")
ax[0].set_title('Original')

# Display the image with increased brightness (img + 100) in the second subplot
ax[1].imshow(imgb, cmap="gray")
ax[1].set_title('img + 100')

# Display the image with increased brightness using cv.add(img, 100) in the third subplot
ax[2].imshow(imgc, cmap="gray")
ax[2].set_title('cv.add')

# Show the plot with all the subplots
plt.show()
```

Increasing the Brightness Using Loops I

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

def image_brighten(image, shift):
    h, w = image.shape
    result = np.zeros(image.shape, dtype=image.dtype)
    for i in range(h):
        for j in range(w):
            # Check for overflow after adding the shift
            no_overflow = True if image[i, j] + shift < 255 else False
            # Apply brightness shift and cap at 255 if there's overflow
            result[i, j] = min(image[i, j] + shift, 255) if no_overflow else 255
    return result

# Read the image in grayscale from the file (path is assumed to be "../images/keira.jpg")
img = cv.imread('images\Sanga-ICC-Hall-of-Fame.jpg', cv.IMREAD_GRAYSCALE)

# Apply the image_brighten function with a brightness shift of 200
imgb = image_brighten(img, 200)

# Create a figure and a set of subplots with 1 row and 2 columns
f, ax = plt.subplots(1, 2)

# Display the original image in the first subplot
ax[0].imshow(img, cmap='gray')
ax[0].set_title('Original')

# Display the image with increased brightness using image_brighten function in the second subplot
ax[1].imshow(imgb, cmap='gray')
ax[1].set_title('image_brighten')

# Show the plot with both subplots
plt.show()
```

Intensity transformation

```python

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

transform = np.arange(0, 256).astype('uint8')

fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $T[f(\mathbf{x})]$')
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_aspect('equal')
#plt.savefig('transform.png')
plt.show()

img_orig = cv.imread('images\Sanga-ICC-Hall-of-Fame.jpg', cv.IMREAD_GRAYSCALE)
print(img_orig.shape)

cv.namedWindow('Image', cv.WINDOW_AUTOSIZE)
cv.imshow('Image', img_orig)
cv.waitKey(0)

image_transformed = cv.LUT(img_orig, transform)
cv.imshow('Image', image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()
```

Intensity Windowing - This used to set the intensity change respect to original image intensity in periticular region (0-100) (100-150) (150-255)

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

c = np.array([(100, 50), (150, 200)])

t1 = np.linspace(0, c[0, 1], c[0, 0] + 1 - 0).astype('uint8')
print(len(t1))

t2 = np.linspace(c[0, 1] + 1, c[1, 1], c[1, 0] - c[0, 0]).astype('uint8')
print(len(t2))

t3 = np.linspace(c[1, 1] + 1, 255, 255 - c[1, 0]).astype('uint8')
print(len(t3))

transform = np.concatenate((t1, t2), axis=0).astype('uint8')
transform = np.concatenate((transform, t3), axis=0).astype('uint8')
print(len(transform))

fig, ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel(r'Input, $f(\mathbf{x})$')
ax.set_ylabel(r'Output, $T[f(\mathbf{x})]$')
ax.set_xlim(0, 255)
ax.set_ylim(0, 255)
ax.set_aspect('equal')
plt.savefig('transform.png')
plt.show()

img_orig = cv.imread('images\Sanga-ICC-Hall-of-Fame.jpg', cv.IMREAD_GRAYSCALE)
cv.namedWindow("Image", cv.WINDOW_AUTOSIZE)
cv.imshow("Image", img_orig)
cv.waitKey(0)

image_transformed = cv.LUT(img_orig, transform)
cv.imshow("Image", image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()
```

Histogram Equilization

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
img = cv.imread('images/vj.jpg', cv.IMREAD_GRAYSCALE)

# Compute histogram and cumulative distribution function (cdf) for the original image
hist, bins = np.histogram(img.ravel(), 256, [0, 256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

# Plot the cumulative distribution function (cdf) and histogram of the original image
plt.plot(cdf_normalized, color='b')
plt.hist(img.flatten(), 256, [0, 256], color='r')
plt.xlim([0, 256])
plt.legend(('cdf', 'histogram'), loc='upper left')
plt.title('Histogram of the Original Image')
plt.show()

# Apply histogram equalization using OpenCV
equ = cv.equalizeHist(img)

# Compute histogram and cumulative distribution function (cdf) for the equalized image
hist, bins = np.histogram(equ.ravel(), 256, [0, 256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max() / cdf.max()

# Plot the cumulative distribution function (cdf) and histogram of the equalized image
plt.plot(cdf_normalized, color='b')
plt.hist(equ.flatten(), 256, [0, 256], color='r')
plt.xlim([0, 256])
plt.legend(('cdf', 'histogram'), loc='upper left')
plt.title('Histogram of the Equalized Image')
plt.show()

# Display the original and equalized images side by side
res = np.hstack((img, equ))
plt.axis('off')
plt.imshow(res, cmap='gray')
plt.show()
```

Gamma correction

```python
%matplotlib inline
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

# Load the original image
img_orig = cv.imread('images/vj.jpg', cv.IMREAD_COLOR)

# Apply gamma correction
gamma = 2
table = np.array([(i/255.0)**gamma*255.0 for i in np.arange(0, 256)]).astype('uint8')
img_gamma = cv.LUT(img_orig, table)

# Convert images to RGB format
img_orig = cv.cvtColor(img_orig, cv.COLOR_BGR2RGB)
img_gamma = cv.cvtColor(img_gamma, cv.COLOR_BGR2RGB)

# Create subplots
f, axarr = plt.subplots(3, 2)

# Display the original and gamma corrected images
axarr[0, 0].imshow(img_orig)
axarr[0, 1].imshow(img_gamma)

# Plot histograms
color = ('b', 'g', 'r')
for i, c in enumerate(color):
    hist_orig = cv.calcHist([img_orig], [i], None, [256], [0, 256])
    axarr[1, 0].plot(hist_orig, color=c)
    
    hist_gamma = cv.calcHist([img_gamma], [i], None, [256], [0, 256])
    axarr[1, 1].plot(hist_gamma, color=c)

# Plot the gamma correction table
axarr[2, 0].plot(table)
axarr[2,0].set_xlabel("Gamma Correction Intensity Transformation")
axarr[2, 0].set_xlim(0, 255)
axarr[2, 0].set_ylim(0, 255)
axarr[2, 0].set_aspect('equal')

axarr[2, 1].axis('off')

# Show the plots
plt.show()
```
